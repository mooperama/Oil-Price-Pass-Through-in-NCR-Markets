```python
# Load the raw Excel sheets
directory_dti = r'data/uncleaned_data/prevailing_prices.xlsx'

# load
df_dti_clean = dp.load_pp(directory_dti)

# standardize
df_dti_clean = dp.add_standardize(df_dti_clean)

# save each cleaned df as a CSV
output_folder = 'data/semi_cleaned_data' # saved to semi_cleaned folder
os.makedirs(output_folder, exist_ok=True)

for year, df in df_dti_clean.items():
    filename = os.path.join(output_folder, f"{year}_cleaned.csv")
    df.to_csv(filename, index=False)
    print(f"Saved {filename}")
```

```python
# path to the folder containing your 2020_cleaned.csv, 2021_cleaned.csv, etc.
CSV_DIR   = "data/semi_cleaned_data"

# path specifically to the Excel file (not just the folder)
XLSX_PATH = "data/uncleaned_data/prevailing_prices.xlsx"

# run the pipeline
panel, reports = dp.run_dti_pipeline(CSV_DIR, XLSX_PATH)

# diagnostics
print(f"Panel Shape: {panel.shape}")
display(panel.head())

# validation report
print("\nCoverage Check")
print(reports["validation"]["coverage"].head())

print(f"\nSpikes repaired: {reports['validation']['n_spikes_repaired'].iloc[0,0]}")
print(f"Bimodal series detected: {len(reports['validation']['bimodal_series'])}")

# Save the final cleaned weekly panel
panel.to_parquet("data/semi_cleaned_data/dti_weekly_panel.parquet", index=False)
panel.to_csv("data/semi_cleaned_data/dti_weekly_panel.csv")
```
